## Frequency exercise

```python
from math import pi, degrees, sqrt, atan
from cmath import polar, exp
```

```python
def c_n(n, A = 1.):
    if n == -2 or n == 2:
        return 1j * A / 4.
    
    return (A * exp(-1j * n * pi)) / (pi * (n^2 - 4))

w_0 = 1. / 2.
```

```python
import pandas as pd

frequency_spectrum = pd.DataFrame(columns = ["n", "nω_0", "|C_n|", "θ_n" ])

for n in range(-3, 5):
    mag, phase = polar(c_n(n))
    
    row = {'n': n, 'nω_0': n * w_0, "|C_n|": round(mag, 3), 'θ_n': round(degrees(phase), 3)}
    frequency_spectrum = frequency_spectrum.append(row, ignore_index=True)


frequency_spectrum
```

```python
plot = frequency_spectrum.plot.bar(x='nω_0', y='|C_n|')
for p in plot.patches:
    x, y = p.get_x(), p.get_height()
    plot.annotate(str(y), xy=(x, y + 0.01))

plot
```

```python
plot = frequency_spectrum.plot.bar(x='nω_0', y='θ_n')
for p in plot.patches:
    x, y = p.get_x(), p.get_height()
    plot.annotate(str(y), xy=(x, y))

plot
```
